### 04. Leia o lucro total (total_profit) de todos os meses e mostre-o usando um gráfico de linha.

O gráfico deve se parecer com esse:

image.png

```
x = sales_df['month_number']
y = sales_df['total_profit']

fig = plt.figure()
fig, ax = plt.subplots()

plt.xticks(x)

ax.set_xlabel('Mês')
ax.set_ylabel('Lucro ($)')
ax.set_title('Lucro obtido por mês')

ax.plot(x, y);
```

### 06. Exiba o número de unidades vendidas por mês para cada produto usando um gráfico multilinha.

image.png

### 10. Exiba o total de vendas de cada produto ao longo dos meses em um gráfico de área. Defina um grau de transparência ao gráfico para deixar a visualização mais agradável.

O gráfico deve ser semelhante a esse:

```
x = sales_df['month_number']

fig = plt.figure()
fig, ax = plt.subplots(figsize=[7, 7])

ax.plot([],[],label='Creme facial', color='blue', linewidth=3)
ax.plot([],[],label='Sabonete de rosto', color='orange', linewidth=3)
ax.plot([],[],label='Pasta de dente', color='green', linewidth=3)
ax.plot([],[],label='Sabonete de banho', color='red', linewidth=3)
ax.plot([],[],label='Shampoo', color='darkorchid', linewidth=3)
ax.plot([],[],label='Hidratante', color='saddlebrown', linewidth=3)

ax.stackplot(x, [sales_df['facecream'], sales_df['facewash'], sales_df['toothpaste'], 
             sales_df['bathingsoap'], sales_df['shampoo'], sales_df['moisturizer']],
             alpha=0.7)

ax.legend(loc='upper left');
```

### 11. Compare o volume de vendas de hidratante e creme facial com um gráfico de barras.

O gráfico de barras deve exibir o número de unidades vendidas por mês para cada produto. Adicione uma barra separada para cada produto no mesmo gráfico. 

Neste mesmo gráfico, apresente o total de produtos vendidos em cada mês. Como a escala de quantidade de produtos é diferente da escala de lucro, crie um eixo gêmeo (pesquise por twin axes).




O gráico deve ser semelhante a esse:

image.png

```
x = sales_df['month_number']

width = 0.35 # The width of the bars

fig = plt.figure()
fig, ax1 = plt.subplots(figsize=[8, 6])

plt.xticks(x)

ax1.set_title('Vendas de hidratante e creme facial')
ax1.set_xlabel('Mês')
ax1.set_ylabel('Unidades Vendidas')
ax1.xaxis.grid(False)
ax1.yaxis.grid(False)

ax1.bar(x - width/2, sales_df['facecream'], width, label='Vendas de Creme Facial')
ax1.bar(x + width/2, sales_df['moisturizer'], width, label='Vendas de Hidratante')

fig.tight_layout()
ax1.legend(loc='upper left')

ax2 = ax1.twinx()

ax2.plot(x, sales_df['total_profit'], color='green', marker='o', markersize=8, linewidth=4, markerfacecolor='white', markeredgewidth=1, markeredgecolor='green')
ax2.xaxis.grid(False)
```

### 12. Crie uma grade de subplots e, para cada painel, exiba um gráfico de dispersão relacionando os dados de vendas de cada produto e o total de lucro obtido no mês. Na última linha da grade deve ser apresentado um único painel com um gráfico de caixas exibindo a distribuição de vendas de cada produto. 


```
x = sales_df['month_number']

data = ["facecream", "facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"]
labels = ['Creme facial', 'Sabonete de rosto', 'Pasta de dente', 
          'Sabonete de banho', 'Shampoo', 'Hidratante']

fig = plt.figure(figsize=[10, 8], constrained_layout=True)
gs = gridspec.GridSpec(3, 3, figure=fig)

ax1 = fig.add_subplot(gs[0, :1])
ax2 = fig.add_subplot(gs[0, 1:2])
ax3 = fig.add_subplot(gs[0, 2:])

ax4 = fig.add_subplot(gs[1, :1])
ax5 = fig.add_subplot(gs[1, 1:2])
ax6 = fig.add_subplot(gs[1, 2:])

ax7 = fig.add_subplot(gs[-1, :])

fig.suptitle("Vendas de Produtos por Mês")

ax1.set_title('Creme facial')
ax1.scatter(x, sales_df['facecream'])

ax2.set_title('Sabonete de rosto')
ax2.scatter(x, sales_df['facewash'])

ax3.set_title('Pasta de dente')
ax3.scatter(x, sales_df['toothpaste'])

ax4.set_title('Sabonete de banho')
ax4.scatter(x, sales_df['bathingsoap'])

ax5.set_title('Shampoo')
ax5.scatter(x, sales_df['shampoo'])

ax6.set_title('Hidratante')
ax6.scatter(x, sales_df['moisturizer'])

ax7.boxplot(sales_df[data].transpose(), labels=labels)

plt.show();
```
